b)список оценок - словарь, где ключами являются предметы
c)Данные студента - кортеж, тк эти данные не изменяются, либо словарь с ключами фио:..., адрес:..., дата рождения:...

```python
def funk(integers):
    dict={}
    for i in range(0,len(integers)):
        list_for_tuple=[]
        for j in range(0,integers[i]):
            check=0
            for k in range(1,j+1):
                if j%k==0:
                    check+=1
            if check==2 and (j in list_for_tuple)==False:
                list_for_tuple.append(j)  
        tuple_for_dict=tuple(list_for_tuple)
        dict[integers[i]]=tuple_for_dict     
    print(dict)
            
a=[143]
funk(a)
```
